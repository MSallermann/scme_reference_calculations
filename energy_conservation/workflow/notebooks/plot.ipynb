```python
import matplotlib.pyplot as plt 
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("results/final.csv")
print(df)
```

```python
E0 = -19.31472938518292
plt.plot(df["timestep"], np.abs(df["energy_tot"] - E0), marker=".")
plt.ylabel(r"$E_\text{final} - E_\text{initial}$ [eV]")
plt.xlabel(r"$\Delta t$ [fs]")
plt.savefig("energy_conservation.png", dpi=300)
```
